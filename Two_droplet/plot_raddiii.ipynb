```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("runs/lam1.0_zeta-4.5/radii_log.csv")

plt.plot(df["time"], df["radius_left"], label="Left droplet")Q
plt.plot(df["time"], df["radius_right"], label="Right droplet")
plt.xlabel("Time")
plt.ylabel("Radius")
plt.title("Droplet Radii vs Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Set path to your simulation outputs
folder = "runs/lam1.0_zeta-4.5"  # adjust as needed
phi0 = np.load(os.path.join(folder, "phi_t00000.npy"))
phiT = np.load(os.path.join(folder, "phi_t08000.npy"))  # assuming 2000/0.0025 = 8000

fig, axs = plt.subplots(1, 2, figsize=(10, 4))
vmin, vmax = -1.0, 1.0  # or use np.min/max(phiT) for autoscaling

im0 = axs[0].imshow(phi0.T, cmap="plasma", origin="lower", vmin=vmin, vmax=vmax)
axs[0].set_title("t = 0")
axs[0].axis("off")

imT = axs[1].imshow(phiT.T, cmap="plasma", origin="lower", vmin=vmin, vmax=vmax)
axs[1].set_title("t = 2000")
axs[1].axis("off")

plt.colorbar(imT, ax=axs, orientation="vertical", shrink=0.8)
plt.tight_layout()
plt.show()
```
