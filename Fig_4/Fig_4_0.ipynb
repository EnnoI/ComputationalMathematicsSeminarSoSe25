```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import label
import os

# ---- Set up domain ----
N = 128
s_start, s_end = -64, 64
x = np.linspace(s_start, s_end, N, endpoint=False)
y = np.linspace(s_start, s_end, N, endpoint=False)
X, Y = np.meshgrid(x, y, indexing="ij")
dx = abs(X[1,0] - X[0,0])

# ---- Initial condition: two droplets ----
def initial_two_droplets(X, Y, R1=10, R2=5, center1=(-20,0), center2=(20,0)):
    phi = -np.ones_like(X)
    mask1 = (X - center1[0])**2 + (Y - center1[1])**2 < R1**2
    mask2 = (X - center2[0])**2 + (Y - center2[1])**2 < R2**2
    phi[mask1] = 1.0
    phi[mask2] = 1.0
    return phi

# ---- FFT-based derivatives (periodic BCs) ----
kx = np.fft.fftfreq(N, d=(s_end - s_start)/N) * 2*np.pi
ky = np.fft.fftfreq(N, d=(s_end - s_start)/N) * 2*np.pi
KX, KY = np.meshgrid(kx, ky, indexing="ij")
K2 = KX**2 + KY**2
K4 = K2**2

# ---- Simulation parameters ----
a, b, eps_val, M = -0.25, 0.25, 1.0, 1.0
tau = 0.01
t_len = 50   # shorter runs to map faster; increase for thoroughness
timesteps = int(t_len / tau)

def run_simulation(lam_val, zeta, phi_0):
    phi = phi_0.copy()
    f_phi = np.fft.fft2(phi)
    for ii in range(timesteps):
        phi3 = phi**3
        dphi_dx = np.fft.ifft2(1j*KX * f_phi).real
        dphi_dy = np.fft.ifft2(1j*KY * f_phi).real
        lap_phi = np.fft.ifft2(-K2 * f_phi).real
        grad_phi2 = dphi_dx**2 + dphi_dy**2
        lap_phi_grad_phi_x = lap_phi * dphi_dx
        lap_phi_grad_phi_y = lap_phi * dphi_dy
        lapl_phi_grad_phi_x_fft = np.fft.fft2(lap_phi_grad_phi_x)
        lapl_phi_grad_phi_y_fft = np.fft.fft2(lap_phi_grad_phi_y)

        nonlinear_term = -K2 * np.fft.fft2(b*phi3 + lam_val*grad_phi2) - 1j * zeta * (KX * lapl_phi_grad_phi_x_fft + KY * lapl_phi_grad_phi_y_fft)
        f_phi = (f_phi + tau*M*nonlinear_term) / (1 + tau*M*(a*K2 + eps_val*K4))
        phi = np.fft.ifft2(f_phi).real
    return phi

# ---- Extract droplet radii from phi field ----
def extract_radii(phi):
    mask = phi > 0
    labeled, num_features = label(mask)
    areas = [np.sum(labeled==i) for i in range(1, num_features+1)]
    if len(areas) == 2:
        R1 = np.sqrt(np.max(areas)*dx**2 / np.pi)
        R2 = np.sqrt(np.min(areas)*dx**2 / np.pi)
        return R1, R2
    return None, None

# ---- Sweep (zeta, lambda) grid and determine regimes ----
zeta_vals = np.linspace(-4, 3, 12)
lambda_vals = np.linspace(-1.5, 1.5, 12)
region_map = np.zeros((len(lambda_vals), len(zeta_vals)))

for i, lam_val in enumerate(lambda_vals):
    for j, zeta in enumerate(zeta_vals):
        print(f"Simulating zeta={zeta}, lambda={lam_val}")
        phi0 = initial_two_droplets(X, Y)
        phi_end = run_simulation(lam_val, zeta, phi0)
        R1_end, R2_end = extract_radii(phi_end)
        if R1_end and R2_end:
            if R2_end > 5:  # small droplet grows
                region_map[i,j] = 2  # region B or C (reverse)
            else:
                region_map[i,j] = 1  # region A (forward)
        else:
            region_map[i,j] = 0  # no clear droplets

# ---- Plot phase diagram ----
plt.figure(figsize=(8,6))
for i, lam_val in enumerate(lambda_vals):
    for j, zeta in enumerate(zeta_vals):
        if region_map[i,j] == 1:
            plt.scatter(zeta, lam_val, c='blue', marker='o', label='Forward' if (i==0 and j==0) else "")
        elif region_map[i,j] == 2:
            plt.scatter(zeta, lam_val, c='red', marker='s', label='Reverse' if (i==0 and j==0) else "")

# Labels
plt.text(-3, 1.2, "A", fontsize=14, color='black')
plt.text(-3, -1, "B", fontsize=14, color='black')
plt.text(2, 1.2, "C", fontsize=14, color='black')

plt.xlabel(r"$\zeta$")
plt.ylabel(r"$\lambda$")
plt.title("Reproduced Phase Diagram - Fig.4(a)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Final AMB+ Simulation Script for Fig. 4(a)
# Modified based on user-provided working code

import sys
import os
import numpy as np
import time
import pyfftw.interfaces as pyfftw
from scipy.ndimage import label
pyfftw.cache.enable()


def inital_two_droplets(X, Y):
    phi = -np.ones_like(X)
    r1 = 10 * (X[1, 0] - X[0, 0])
    r2 = 5 * (X[1, 0] - X[0, 0])
    mask1 = (X + 0.3)**2 + Y**2 < r1**2
    mask2 = (X - 0.3)**2 + Y**2 < r2**2
    phi[mask1] = 1.0
    phi[mask2] = 1.0
    return phi


def track_droplet_sizes(phi):
    threshold = 0.0
    labeled, num = label(phi > threshold)
    sizes = [np.sum(labeled == i) for i in range(1, num + 1)]
    if len(sizes) >= 2:
        sizes = sorted(sizes)[-2:]
        r1 = np.sqrt(sizes[1] / np.pi)
        r2 = np.sqrt(sizes[0] / np.pi)
        return r1, r2
    return 0.0, 0.0


def solve_ambplus_2D(phi_0=None, c_0=0.4, t_state=0.0, t_len=100.0, tau=0.01, eps_val=1., a=-0.25, b=0.25, lam_val=1.75, zeta=2.0, D=0.0, M=1., s_start=-0.5, s_end=0.5, s_N=200, tag="sim"):
    log_file = f"{tag}_log.csv"
    radii_file = f"{tag}_radii.csv"
    prev_iter = 0

    # Space discretization
    L = s_end - s_start
    x = np.linspace(s_start, s_end, s_N, endpoint=False)
    y = np.linspace(s_start, s_end, s_N, endpoint=False)
    X, Y = np.meshgrid(x, y, indexing="ij")
    kx = np.fft.fftfreq(s_N, d=L/s_N)*2*np.pi
    ky = np.fft.fftfreq(s_N, d=L/s_N)*2*np.pi
    KX, KY = np.meshgrid(kx, ky, indexing="ij")
    K_2 = KX**2 + KY**2
    K_4 = K_2**2

    dealiasing_mask = K_2 < (0.67 * K_2.max())

    dx = abs(X[1][0] - X[0][0])
    dy = abs(Y[0][1] - Y[0][0])
    dS = dx*dy

    if phi_0 is None:
        phi = inital_two_droplets(X, Y)
    else:
        phi = phi_0

    f_phi = pyfftw.numpy_fft.fft2(phi, norm="backward", threads=8)

    t_N = round(t_len/tau)
    gaussian_scale = np.sqrt(tau / dS)

    if not os.path.exists(log_file):
        with open(log_file, 'w') as f:
            f.write("iteration,time,total_mass,phi_min,phi_max\n")
    if not os.path.exists(radii_file):
        with open(radii_file, 'w') as f:
            f.write("iteration,time,radius1,radius2\n")

    start = time.time()
    for ii in range(prev_iter, prev_iter+t_N):
        if ii % 100 == 0:
            end = time.time()
            r1, r2 = track_droplet_sizes(phi)
            print(f"t = {ii * tau:.2f}, r1 = {r1:.4f}, r2 = {r2:.4f}")
            with open(radii_file, 'a') as f:
                f.write(f"{ii},{ii*tau},{r1},{r2}\n")
            with open(log_file, 'a') as f:
                f.write(f"{ii},{ii*tau},{np.sum(phi.real)*dS / L**2},{np.min(phi.real)},{np.max(phi.real)}\n")
            start = time.time()

        phi_3 = phi**3
        dphi_dx = pyfftw.numpy_fft.ifft2(1j * KX * f_phi, norm="backward", threads=8)
        dphi_dy = pyfftw.numpy_fft.ifft2(1j * KY * f_phi, norm="backward", threads=8)

        laplacian_phi = pyfftw.numpy_fft.ifft2(-K_2 * f_phi, norm="backward", threads=8)
        lapl_phi_prod_grad_phi_x_fft = pyfftw.numpy_fft.fft2(laplacian_phi * dphi_dx, norm="backward", threads=8)
        lapl_phi_prod_grad_phi_y_fft = pyfftw.numpy_fft.fft2(laplacian_phi * dphi_dy, norm="backward", threads=8)

        grad_phi_2 = dphi_dx**2 + dphi_dy**2

        non_linear_term = - K_2 * pyfftw.numpy_fft.fft2(b * phi_3 + lam_val * grad_phi_2, norm="backward", threads=8)
        non_linear_term -= 1j * zeta * (KX * lapl_phi_prod_grad_phi_x_fft + KY * lapl_phi_prod_grad_phi_y_fft)
        non_linear_term *= dealiasing_mask

        f_phi = (f_phi + tau * M * non_linear_term) / (1. + tau * M *(a * K_2 + eps_val * K_4))
        phi = pyfftw.numpy_fft.ifft2(f_phi, norm="backward", threads=8)
        t_state += tau

    np.save(f"{tag}_final.npy", phi.real)
    print(f"Simulation complete: {tag}_final.npy saved.")


def main():
    np.random.seed(0)
    solve_ambplus_2D(phi_0=None, c_0=0.6, s_N=128, tau=0.02, t_len=10000, D=0.0, zeta=-4.0, lam_val=-1.0, s_start=-0.5, s_end=0.5, tag="fig4a")

if __name__ == "__main__":
    main()
```

```python
# This version modifies your original AMB+ simulation code to:
# - Create a two-droplet or two-bubble initial condition
# - Track radii over time
# - Compare their dynamics under different (zeta, lambda) for Region A, B, C

import numpy as np
import os
import pyfftw.interfaces as pyfftw
import time
from scipy.ndimage import label
pyfftw.cache.enable()

def initial_c_0_2D(X, Y, c_0=0.5, init_noise=0.001):
    return (2.*c_0 - 1.) + init_noise * np.random.randn(*X.shape)

def create_two_droplets(X, Y, r=0.05, centers=[(-0.3, 0), (0.3, 0)], c_in=0.8, c_out=0.0):
    phi = initial_c_0_2D(X, Y, c_0=c_out, init_noise=0.001)
    for cx, cy in centers:
        mask = (X - cx)**2 + (Y - cy)**2 < r**2
        phi[mask] = 2. * c_in - 1.
    return phi

def create_two_bubbles(X, Y, r=0.05, centers=[(-0.3, 0), (0.3, 0)], c_in=0.1, c_out=0.9):
    phi = initial_c_0_2D(X, Y, c_0=c_out, init_noise=0.001)
    for cx, cy in centers:
        mask = (X - cx)**2 + (Y - cy)**2 < r**2
        phi[mask] = 2. * c_in - 1.
    return phi

def track_droplet_sizes(phi, threshold=0.5):
    labeled, num = label(phi > threshold)
    sizes = []
    for i in range(1, num + 1):
        area = np.sum(labeled == i)
        radius = np.sqrt(area / np.pi)
        sizes.append(radius)
    return sorted(sizes)[-2:] if len(sizes) >= 2 else [0, 0]

def solve_case(zeta, lam_val, out_prefix, init_mode="droplet", t_len=10000, tau=0.1, N=256):
    L = 1.0
    x = np.linspace(-0.5*L, 0.5*L, N, endpoint=False)
    y = np.linspace(-0.5*L, 0.5*L, N, endpoint=False)
    X, Y = np.meshgrid(x, y, indexing='ij')
    dx = x[1] - x[0]
    KX, KY = np.meshgrid(np.fft.fftfreq(N, d=dx)*2*np.pi, np.fft.fftfreq(N, d=dx)*2*np.pi, indexing="ij")
    K2 = KX**2 + KY**2
    K4 = K2**2

    dealias = K2 < (0.67 * K2.max())

    if init_mode == "droplet":
        phi = create_two_droplets(X, Y, r=0.08, c_in=0.9, c_out=0.1)
    elif init_mode == "bubble":
        phi = create_two_bubbles(X, Y, r=0.08, c_in=0.1, c_out=0.9)
    else:
        raise ValueError("Unknown init_mode")

    f_phi = pyfftw.numpy_fft.fft2(phi)

    a, b, eps, D, M = -0.25, 0.25, 1.0, 0.0, 1.0
    t_state = 0.0
    t_N = int(t_len / tau)
    radii_log = []

    for step in range(t_N):
        phi3 = phi**3
        dphi_dx = pyfftw.numpy_fft.ifft2(1j * KX * f_phi)
        dphi_dy = pyfftw.numpy_fft.ifft2(1j * KY * f_phi)
        grad_phi2 = dphi_dx.real**2 + dphi_dy.real**2

        lap_phi = pyfftw.numpy_fft.ifft2(-K2 * f_phi)
        lap_phi_grad_x = pyfftw.numpy_fft.fft2(lap_phi * dphi_dx)
        lap_phi_grad_y = pyfftw.numpy_fft.fft2(lap_phi * dphi_dy)

        nonlinear = -K2 * pyfftw.numpy_fft.fft2(b * phi3 + lam_val * grad_phi2) \
                    -1j * zeta * (KX * lap_phi_grad_x + KY * lap_phi_grad_y)
        nonlinear *= dealias

        f_phi = (f_phi + tau * M * nonlinear) / (1 + tau * M * (a * K2 + eps * K4))
        phi = pyfftw.numpy_fft.ifft2(f_phi).real

        if step % 100 == 0:
            thresh = 0.0 if init_mode == "droplet" else -0.2
            r1, r2 = track_droplet_sizes(phi, threshold=thresh)
            radii_log.append((t_state, r1, r2))
            print(f"step {step}, t={t_state:.2f}, radii: {r1:.4f}, {r2:.4f}")


        t_state += tau

    np.savetxt(f"{out_prefix}_radii.csv", np.array(radii_log), delimiter=",", header="time,radius1,radius2", comments="")
    np.save(f"{out_prefix}_final.npy", phi)

if __name__ == "__main__":
    solve_case(zeta=-1.0, lam_val=0.5, out_prefix="regionA", init_mode="droplet")  # Region A
    solve_case(zeta=-4.0, lam_val=-1.0, out_prefix="regionB", init_mode="droplet") # Region B
    solve_case(zeta=4.0, lam_val=1.0, out_prefix="regionC", init_mode="bubble")   # Region C
```
